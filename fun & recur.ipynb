```python
# program to find sum of multiple numbers 

def find_sum(*numbers):
    result = 0
    
    for num in numbers:
        result = result + num
    
    print("Sum = ", result)

# function call with 3 arguments
find_sum(1, 2, 3)

# function call with 2 arguments
find_sum(4, 9)
```

```python
def mul(a,b):
    if b == 1:
        return a
    else:
        return a + mul(a,b-1)
    
print(mul(4,3))
```

```python
def G_D_C(a, b):
	low = min(a, b)
	high = max(a, b)

	if low == 0:
		return high
	elif low == 1:
		return 1
	else:
		return G_D_C(low, high%low)
	
a = int(input('Enter the value of a: '))
b = int(input('Enter the value of b: '))
print(G_D_C(a,b))
```
